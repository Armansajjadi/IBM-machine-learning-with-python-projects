# trying K-means, hirerichal, DBSCAN on Customer dataset

```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
```

```python
customerData = pd.read_csv("Customer.csv")

customerData.head()
```

```python
customerData["Gender-num"] = customerData["Gender"].map({'Male':1,'Female':0})

myDataAsArray = np.asarray(customerData[["Gender-num", "Age", "Annual Income (k$)", "Spending Score (1-100)"]])
```

```python
normalizedData = preprocessing.StandardScaler().fit_transform(myDataAsArray)
```

## K-Means

```python
silhouette_scores = {}

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(normalizedData)
    score = silhouette_score(normalizedData, labels)
    silhouette_scores[k] = score

# Find best k
best_k = max(silhouette_scores, key=silhouette_scores.get)

kMeansModel=KMeans(n_clusters=best_k, random_state=42)
kMeansModel.fit(normalizedData)
labels = kMeansModel.labels_
labels
```

```python
customerData["Kmeans_cluster"] = labels
customerData.head()
```

## DBSCAN

```python
dbscan = DBSCAN(eps=0.7, min_samples=3)
labels = dbscan.fit_predict(normalizedData)

customerData['dbscan_cluster'] = labels
customerData['dbscan_cluster'].value_counts()
```

```python
plt.figure(figsize=(8, 6))

mask = customerData['dbscan_cluster'] != -1
plt.scatter(customerData.loc[mask, "Age"],
            customerData.loc[mask, "Annual Income (k$)"],
            c=customerData.loc[mask, "dbscan_cluster"],
            cmap='viridis',
            label='Clusters')

noise = customerData['dbscan_cluster'] == -1
plt.scatter(customerData.loc[noise, "Age"],
            customerData.loc[noise, "Annual Income (k$)"],
            c='red',
            marker='x',
            label='Noise')

plt.title("DBSCAN Clustering")
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
plt.legend()
plt.grid(True)
plt.show()
```
